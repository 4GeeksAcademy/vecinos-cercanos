```python
# Your code here

#carag de data set

import pandas as pd

movies = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv")
credits = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv")

movies.head()
```

```python
credits.head()
```

```python
#creacion de la base de datos 
import sqlite3

# coneccion con SQlite
conn = sqlite3.connect('TMDB.db')

# Guardoel csv separado en tablas
movies.to_sql('movies', conn, if_exists='replace', index=False)
credits.to_sql('credits', conn, if_exists='replace', index=False)
```

```python
# junto las 2 tablas de movies y credits:
join_query = """
SELECT c.movie_id, c.title, m.overview, m.genres, m.keywords, c.cast, c.crew
FROM movies m
INNER JOIN credits c ON m.title = c.title
"""

unified_table = pd.read_sql_query(join_query, conn)
unified_table.head()
```

```python
# transformaciÃ³n de los datos:
import json

# sacamos los nombres de json:
def extract_names(json_str, num_names=None):
    try:
        json_list = json.loads(json_str)
        names = [item['name'] for item in json_list]
        if num_names is not None:
            names = names[:num_names]
        return names
    except json.JSONDecodeError:
        return []

# creamos el csv:
unified_table = pd.DataFrame(unified_table)

# Applying the function to the columns
unified_table['genres'] = unified_table['genres'].apply(lambda x: ', '.join(extract_names(x)))
unified_table['keywords'] = unified_table['keywords'].apply(lambda x: ', '.join(extract_names(x)))
unified_table['cast'] = unified_table['cast'].apply(lambda x: ', '.join(extract_names(x, num_names=3)))
unified_table['crew'] = unified_table['crew'].apply(lambda x: ', '.join([crew_member['name'] for crew_member in json.loads(x) if crew_member['job'] == 'Director']))
unified_table['overview'] = unified_table['overview'].apply(lambda x: [x])

print("\nUnified table:")
unified_table.head()
```

```python
# funciÃ³n para convertir en cadena los elementos y poder unirlos

def join_and_convert(lst):
    return ''.join(map(str, lst))

# Aplicando la funciÃ³n para convertir elementos en las columnas especificadas en cadenas

columns_to_process = ['overview', 'genres', 'keywords', 'cast', 'crew']
for column in columns_to_process:
    unified_table[column] = unified_table[column].apply(join_and_convert)

# Concatenando las columnas procesadas en una nueva columna 'etiquetas':

unified_table['tags'] = unified_table['overview'] + unified_table['genres'] + unified_table['keywords'] + unified_table['cast'] + unified_table['crew']
unified_table["tags"] = unified_table["tags"].apply(lambda x: ",".join(x).replace(",", ""))
unified_table["tags"] = unified_table["tags"].apply(lambda x: '"'.join(x).replace('"', ""))
unified_table.drop(columns = ["genres", "keywords", "cast", "crew", "overview"], inplace = True)
unified_table['tags']
```

```python
#construyo el modelo:

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

vectorizer = TfidfVectorizer()
text_vectors = vectorizer.fit_transform(unified_table["tags"])

model = NearestNeighbors(n_neighbors = 6, algorithm = "brute", metric = "cosine")
model.fit(text_vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(text_vectors)

def recommend(movie):
    movie_index = unified_table[unified_table["title"] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse = True , key = lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(unified_table.iloc[i[0]].title)

print("Pelicula recomendada:")
recommend("Spectre")
```
